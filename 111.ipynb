```python
import pandas as pd
df_pom=pd.read_csv('1582.csv')
df_pom
```

```python
df_pom.columns
```

```python
# Добавление нового столбца с первыми 13 символами
df_pom['Муниципальный район'] = df_pom['Кадастровый номер'].str[:5]

# Группировка и подсчет количества
#grouped = df_pom.groupby(['Муниципальный район']).sum('Кад. стоимость')
grouped = df_pom.groupby('Муниципальный район')['Кад. стоимость'].agg( ['sum', 'count'] )
grouped =grouped.rename(columns={
    'sum': 'КС сумма',
    'count': 'Количество ОН',
}).reset_index()

grouped['КС сумма млн.руб']=grouped['КС сумма']/1000000

# Просмотр результата
print(grouped)
```

```python
# Сохранение результата
grouped.to_csv('C:\Afbks\grouped_мун_кс.csv')

# Просмотр результата
print(grouped)
```
