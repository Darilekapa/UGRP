```python
import pandas as pd

# Чтение конкретного листа и сохранение в CSV
def excel_to_csv_pandas(excel_file, sheet_name, csv_file):
    """
    Конвертирует указанный лист Excel в CSV
    
    Args:
        excel_file (str): путь к Excel файлу
        sheet_name (str): имя листа для конвертации
        csv_file (str): путь для сохранения CSV
    """
    try:
        # Читаем конкретный лист
        df = pd.read_excel(excel_file, sheet_name=sheet_name)
        # Сохраняем в CSV
        df.to_csv(csv_file, index=False, encoding='utf-8')
        print(f"Успешно конвертирован лист '{sheet_name}' в {csv_file}")
    except Exception as e:
        print(f"Ошибка: {e}")

# Использование
excel_to_csv_pandas('EGRN_НН.xlsx', 'Sheet1', 'pom1.csv')
```

```python
excel_to_csv_pandas('EGRN_НН.xlsx', 'Sheet2', 'pom2.csv')
```

```python
import pandas as pd
df_pom=pd.read_csv('pom1.csv')
df_pom1=pd.read_csv('pom2.csv')
```

```python
combined_vertical = pd.concat([df_pom, df_pom1]) 
combined_vertical
```

```python
#combined_vertical.columns
```

```python
#combined_vertical.info()
```

```python
# Добавление нового столбца с первыми 13 символами
combined_vertical['Муниципальный район'] = combined_vertical['Кадастровый номер'].str[:5]

# Группировка и подсчет количества
#grouped = df_pom.groupby(['Муниципальный район']).sum('Кад. стоимость')
grouped = combined_vertical.groupby('Муниципальный район')['Кад. стоимость'].agg( ['sum', 'count'] )
grouped =grouped.rename(columns={
    'sum': 'КС сумма',
    'count': 'Количество ОН',
}).reset_index()

grouped['КС сумма млн.руб']=grouped['КС сумма']/1000000

# Просмотр результата
print(grouped)
```

```python
# Сохранение результата
grouped.to_csv('C:\Afbks\grouped_мун_кс_pom.csv')

# Просмотр результата
print(grouped)
```
